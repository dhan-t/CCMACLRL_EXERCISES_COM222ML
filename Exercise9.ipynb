```python
# Create a LabelEncoder object
encoder = LabelEncoder()

# Iterate over each column in the DataFrame
for column in df.columns:
    # Check if the column data type is 'object' (indicating categorical data)
    if df[column].dtype == 'object':
        # Fit the encoder on the unique values in the column
        encoder.fit(df[column])
        # Transform the column values to numerical labels
        df[column] = encoder.transform(df[column])
```

```python
# Array of features
X = df.drop(columns=['loan_status']).values

# Array of Labels
y = df['loan_status'].values
```

```python
# put your answer here
X_train, X_test, y_train, y_test = train_test_split(X,  y, test_size=0.2, random_state=5)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train, y_train)
```

- Perform cross validation

```python
# put your answer here
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn, X, y, cv=5)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
score_list['knn'] = scores.mean()
```
